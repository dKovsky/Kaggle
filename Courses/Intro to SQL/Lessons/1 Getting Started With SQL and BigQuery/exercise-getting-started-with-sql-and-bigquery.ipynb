### 2) Explore the table schema

How many columns in the `crime` table have `TIMESTAMP` data?

```python
# Reference to the "crime" table
table_ref = dataset_ref.table("crime")

# API request - fetch the table schema
table = client.get_table(table_ref)

# List all columns and their data types
columns = table.schema

# Count the number of columns with TIMESTAMP data type
num_timestamp_columns = sum(1 for column in columns if column.field_type == 'TIMESTAMP')

# Print the result
print("Number of columns in the 'crime' table with TIMESTAMP data:", num_timestamp_columns)
```
